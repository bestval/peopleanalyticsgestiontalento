```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
X, y = make_regression(n_features=1, n_samples=500, random_state=42, noise=10)
plt.xlabel('Feature X')
plt.ylabel('Target y')
plt.scatter(X, y, s=5)
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X)
y_pred
```

```python
plt.scatter(X, y, s=25, label='training')
plt.scatter(X, y_pred, s=25, label='predictions')
plt.xlabel('Feature X')
plt.ylabel('Target y')
plt.legend()
```
