```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Cargar dataset de ejemplo (Iris)
iris = load_iris()
X = iris.data
y = iris.target

# Entrenar un rbol de decisin
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X, y)

# Dibujar el rbol
plt.figure(figsize=(12, 8))
plot_tree(clf, 
          feature_names=iris.feature_names, 
          class_names=iris.target_names, 
          filled=True, 
          rounded=True)
plt.title("rbol de decisin (Iris Dataset)")
plt.show()
```

 Estructura de un nodo

Cada caja (nodo) del rbol tiene la siguiente informacin:

Condicin: como petal length (cm) <= 2.45

 gini: ndice de impureza (entre 0 y 0.5 para 2 clases; mximo 0.667 para 3 clases).

 samples: cuntos datos llegan a ese nodo.

 value: distribucin de clases [setosa, versicolor, virginica]

 class: clase dominante en ese nodo.

El rbol comienza dividiendo por longitud del ptalo.

Si es menor o igual a 2.45  izquierda (setosa).

Si es mayor  derecha (versicolor o virginica).


    gini = 0.0
    samples = 50
    value = [50, 0, 0]
    class = setosa

Esto indica que esos 50 ejemplos son 100% setosa.

    gini = 0.444
    samples = 6
    value = [0, 2, 4]
    class = virginica

Aqu hay mezcla: 2 versicolor y 4 virginica  todava se predice virginica, pero hay incertidumbre (por eso el gini  0.444). 

 Cmo interpreta el rbol

El rbol toma decisiones por pasos, dividiendo el espacio en funcin de los valores de caractersticas (como petal length y petal width). Al final, asigna una clase segn mayora en esa hoja.

 Qu significa gini?

El ndice Gini mide la impureza de un nodo:

Gini = 0  nodo puro (todos de la misma clase).

Gini  0.5 (o ms en 3 clases)  clases mezcladas.
