```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Cargar dataset de ejemplo (Iris)
iris = load_iris()
X = iris.data
y = iris.target

# Entrenar un árbol de decisión
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X, y)

# Dibujar el árbol
plt.figure(figsize=(12, 8))
plot_tree(clf, 
          feature_names=iris.feature_names, 
          class_names=iris.target_names, 
          filled=True, 
          rounded=True)
plt.title("Árbol de decisión (Iris Dataset)")
plt.show()
```

🔢 Estructura de un nodo

Cada caja (nodo) del árbol tiene la siguiente información:

Condición: como petal length (cm) <= 2.45

➡️ gini: índice de impureza (entre 0 y 0.5 para 2 clases; máximo 0.667 para 3 clases).

➡️ samples: cuántos datos llegan a ese nodo.

➡️ value: distribución de clases [setosa, versicolor, virginica]

➡️ class: clase dominante en ese nodo.

El árbol comienza dividiendo por longitud del pétalo.

Si es menor o igual a 2.45 → izquierda (setosa).

Si es mayor → derecha (versicolor o virginica).


    gini = 0.0
    samples = 50
    value = [50, 0, 0]
    class = setosa

Esto indica que esos 50 ejemplos son 100% setosa.

    gini = 0.444
    samples = 6
    value = [0, 2, 4]
    class = virginica

Aquí hay mezcla: 2 versicolor y 4 virginica → todavía se predice virginica, pero hay incertidumbre (por eso el gini ≈ 0.444). 

🎯 Cómo interpreta el árbol

El árbol toma decisiones por pasos, dividiendo el espacio en función de los valores de características (como petal length y petal width). Al final, asigna una clase según mayoría en esa hoja.

📌 ¿Qué significa gini?

El índice Gini mide la impureza de un nodo:

Gini = 0 → nodo puro (todos de la misma clase).

Gini ≈ 0.5 (o más en 3 clases) → clases mezcladas.
