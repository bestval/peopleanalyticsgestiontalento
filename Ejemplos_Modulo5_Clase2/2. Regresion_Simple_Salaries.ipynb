```python
import pandas as pd # Para manejo de datos
from sklearn.model_selection import train_test_split # Para dividir los datos de entrenamiento y prueba
import matplotlib.pyplot as plt # Para gráficos
from sklearn.linear_model import LinearRegression # Modelo de regresión lineal de scikit-learn
```

Cargar el archivo CSV que contiene datos de años de experiencia y salarios

```python
df = pd.read_csv("../data/salaries_experience.csv")
```

df.head() # Cargar las primeras 5 filas

```python
df.head()
```

df.shape mostrar tamaño

```python
df.shape
```

df.describe() para los estadísticos

```python
df.describe()
```

```python
X = df.iloc[:, :-1]

X
```

```python
y = df.iloc[:, -1]

y
```

Division de los datos del modelo entre entrenamiento y testeo

```python
X_train, X_test, y_traing, y_test = train_test_split(X, y, test_size = 0.20, random_state=42)
```

```python
plt.scatter(X_test, y_test, c='orange')
plt.scatter(X_train, y_traing)
plt.title('Datos')
plt.xlabel('YearExperience')
plt.ylabel('Salary')
plt.show()
```

Entrenamiento del modelo con regresion lineal con los datos de entrenamiento

```python
model = LinearRegression()

model.fit(X_train, y_traing) # conjunto de entrenamiento
```

Se usan los datos de prueba (X_test) para que el modelo prediga los salarios

```python
y_pred = model.predict(X_test) #conjunto de validación, predice sobre datos que no conoce
```

```python
plt.scatter(X_test, y_test, c='orange')

plt.scatter(X_train, y_traing)

plt.plot(X_test, y_pred, c='b')

plt.title('Datos')
plt.xlabel('YearExperience')
plt.ylabel('Salary')

plt.show()
```

```python
model.predict([[9.2]])
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

r2 = r2_score(y_test, y_pred)

mae = mean_absolute_error(y_test, y_pred)

print('r2:', r2)
print('mae:', mae)
```
