```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
X, y = make_regression(n_features=1, n_samples=500, random_state=42, noise=10)
```

```python
plt.xlabel('Feature X')
plt.ylabel('Target y')
plt.scatter(X, y, s=5)
```

```python
model = LinearRegression()

model.fit(X, y)
```

```python
model.coef_
model.intercept_ 
```

```python
y_pred = model.predict(X)
```

```python
plt.scatter(X, y, s=25, label='training')
plt.scatter(X, y_pred, s=25, label='predictions')
plt.xlabel('Feature X')
plt.ylabel('Target y')
plt.legend()
```
